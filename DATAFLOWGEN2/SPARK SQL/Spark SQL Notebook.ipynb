```sql
CREATE TABLE IF NOT EXISTS sales (
    order_id              STRING,
    order_date            DATE,
    customer_id           STRING,
    product_id            STRING,
    quantity              INT,
    unit_price            DOUBLE,
    discount_percent      DOUBLE,
    gross_sales           DOUBLE
)
```

```sql
UPDATE sales
SET discount_percent = 15.0,
  gross_sales = qty * unit_price * (1 - 15.0/100.0)
WHERE order_date BETWEEN DATE '2025-08-01' AND DATE '2025-08-04';
```
