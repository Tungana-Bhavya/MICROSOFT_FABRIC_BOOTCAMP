 #### Peforming inner Join between fact_sale table and dimension_stock_item and Retrieving top 1000 rows

```sql
SELECT TOP 1000 i.StockItem , f.*
FROM [Laketbhavya0541].[dbo].[fact_sale] f
INNER JOIN [Laketbhavya0541].[dbo].[dimension_stock_item] i
ON f.StockItemKey = i.StockItemKey
```

#### Top 10 Stock Item

```sql
SELECT TOP 10 i.StockItem, COUNT(*) Count, SUM(f.Quantity) Quantity, SUM(f.Quantity * f.UnitPrice) Gross
FROM Laketbhavya0541.dbo.fact_sale f
INNER JOIN Laketbhavya0541.dbo.dimension_stock_item i
ON f.StockItemKey = i.StockItemKey
GROUP BY i.StockItem ORDER BY 4 DESC
```

#### Retrieving Customer name and BillToCustomer

```sql
 SELECT DISTINCT c1.Customer, c1.BillToCustomer, c2.BillToCustomerKey
 FROM Laketbhavya0541.dbo.dimension_customer c1
 JOIN Laketbhavya0541.dbo.fact_sale c2
 ON c1.CustomerKey = c2.CustomerKey
```
